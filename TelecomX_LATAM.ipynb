```python
import pandas as pd
import requests
import json
```

```python
url = ('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

```python
response = requests.get(url)
data = response.json()
telecom = pd.DataFrame(data)
```

```python
telecom
```

```python
telecom = pd.json_normalize(data)
telecom.head()
```

```python
telecom.columns
```

```python
telecom = pd.json_normalize(data)
telecom.rename(columns={'account.Charges.Monthly': 'Charges_Monthly',
                        'account.Charges.Total': 'Charges_Total'}, inplace=True)
telecom.rename(columns=lambda x: x.split('.')[-1], inplace=True)
telecom.sample(11)
```

```python
telecom.info()
```

```python
for col in telecom.columns:
    print(f"valores unicos en la columna '{col}':{telecom[col].nunique()}")
    if telecom[col].nunique() < 50:
       print(telecom[col].unique())
       print('-' * 50)
```

```python
vacios = telecom.apply(lambda x: x.astype(str).str.strip() =='').sum()
print('\nValores en blanco o string vacios:')
print(vacios[vacios > 0])
```
