```python
import pandas as pd
import requests
import json
```

```python
url = ('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

```python
response = requests.get(url)
data = response.json()
telecom = pd.DataFrame(data)
print(telecom.head)
```

```python
telecom = pd.json_normalize(data)
telecom.head()
```

```python
telecom.columns
```

```python
telecom = pd.json_normalize(data)
telecom.rename(columns={'account.Charges.Monthly': 'Charges_Monthly',
                        'account.Charges.Total': 'Chages_Total'}, inplace=True)
telecom.rename(columns=lambda x: x.split('.')[-1], inplace=True)
telecom.head()
```

```python
telecom.info()
```

```python
import numpy as np
```

```python
print(telecom.isnull().sum())
print("\n")
```

```python
valores_vacios = telecom.apply(lambda x: x.astype(str).str.strip()=="").sum()
print("\nValores vacios:")
print(valores_vacios[valores_vacios > 0])
```

#ðŸ”§ TransformaciÃ³n

```python
telecom.sample(10)
```

```python
col_yn  = ['Churn','Partner','Dependents','PhoneService','PaperlessBilling']
telecom[col_yn].sample(6)
```

```python
telecom.loc[:,col_yn] = telecom[col_yn].astype(str)
telecom.loc[:,col_yn] = telecom.loc[:, col_yn].replace({'Yes':1,'No':0, '': np.nan})
telecom.loc[:,col_yn] = telecom.loc[:, col_yn].fillna(0)
telecom[col_yn].sample(6)
```

```python
telecom[col_yn] = telecom[col_yn].astype(np.int64)
telecom.info()
```

```python
telecom['Chages_Total'] = pd.to_numeric(telecom['Chages_Total'], errors='coerce')
telecom.info()
```

```python
telecom['Chages_Total'] = telecom['Chages_Total'].fillna(0)
```
